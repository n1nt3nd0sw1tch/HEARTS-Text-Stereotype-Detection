```python
# Set a fixed seed
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
```

```python

print("CUDA available:", torch.cuda.is_available())
print("Device name:", torch.cuda.get_device_name(0))
```

```python
script_path = Path("/Users/rinlobachevskii/Desktop/COMP0173/HEARTS-Text-Stereotype-Detection/Model Training and Evaluation/BERT_Models_Fine_Tuning.py")
hearts_models = SourceFileLoader("hearts_models", str(script_path)).load_module()

# Call the functions
train_mgsd, test_mgsd = hearts_models.data_loader("MGSD.csv", "stereotype", "MGSD", 1_000_000, 5)
model_dir = hearts_models.train_model(train_mgsd, "albert/albert-base-v2", 64, 3, 2e-5, "model_output_albertv2", "mgsd_baseline", 42)
report = hearts_models.evaluate_model(test_mgsd, model_dir, "results_albertv2", "mgsd", 42)
report
```
